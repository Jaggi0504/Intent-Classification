```python
index = pc.Index("flowlly")

index.upsert(
    vectors=[
        {
            "id": "vec1", 
            "values": schedule_update_response.data[0].embedding, 
            "metadata": {"intent": "schedule_update"}
        }, {
            "id": "vec2", 
            "values": task_create_response.data[0].embedding, 
            "metadata": {"intent": "task_create"}
        }, {
            "id": "vec3", 
            "values": query_response.data[0].embedding, 
            "metadata": {"intent": "query"}
        },
    ],
    namespace= "flowlly_ns"
)
```
